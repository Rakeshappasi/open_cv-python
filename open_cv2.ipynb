```python
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

```

```python
image = cv2.imread('Happy_Goku.jpg')
plt.imshow(image)
```

```python
print(image.shape)
```

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print(gray_image)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.title("Gray image")
plt.imshow(gray_image)

plt.subplot(1,2,2)
plt.title("org image")
plt.imshow(image)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.title("Gray image")
plt.imshow(cv2.cvtColor(gray_image,cv2.COLOR_BGR2RGB))

plt.subplot(1,2,2)
plt.title("org image")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))


```

```python
b,g,r = cv2.split(image)
#cv2_imshow(b)
print(b)
```

```python
zeros = np.zeros((image.shape[0],image.shape[1]), np.uint8)
blue = cv2.merge([b,zeros ,zeros])
green= cv2.merge([zeros , g ,zeros])
red= cv2.merge([zeros ,zeros ,r])
cv2_imshow(blue)
cv2_imshow(green)
cv2_imshow(red)
```

```python
c_image = cv2.merge([b,g,r])
cv2_imshow(c_image)
```

```python
canvas= np.zeros([512,512,3],np.uint8)
cv2.line(canvas,(0,0),(511,511),(255,0,0),(3))
cv2_imshow(canvas)
```

```python
cap= cv2.VideoCapture("01.Introduction to Data Structures & Algorithms.mp4")

while cap.isOpened():
    sucess,frame = cap.read()
    if sucess:
        gray= cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow("grey",gray)
        cv2.imshow("Frame",frame)
    else:
        break
```
