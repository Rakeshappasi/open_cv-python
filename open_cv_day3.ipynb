```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("Happy_Goku.jpg")
height,width,channel = image.shape
```

```python
#matrix
m = np.float32([[1,0,50],[0,1,50]])
transalted_image = cv2.warpAffine(image,m,(height,width))
cv2_imshow(image)
cv2_imshow(transalted_image)
```

```python
#rotation
#matrix
m = np.float32([[1,0,50],[0,1,50]])
r_m= cv2.getRotationMatrix2D((height/2,width/2) , 90, 1)
transalted_image = cv2.warpAffine(image,m,(height,width))
r_i= cv2.warpAffine(image,r_m,(height,width))
scaled_image = cv2.resize(image, None , fx=5 ,fy=3)
#cv2_imshow(image)
#cv2_imshow(transalted_image)
#cv2_imshow(r_i)
cv2_imshow(scaled_image )
```

```python
#blending
image_2 = cv2.imread("R.jfif")
image_2 = cv2.resize(image_2 ,(image.shape[1],image.shape[0]))
added = image + image_2
blend= cv2.addWeighted(image,0.8,image_2,0.2,gamma=0.1)
#cv2_imshow(added)
cv2_imshow(blend)
```

```python
#convolution
 
kernel = np.float32([[100,110,0],
                     [0,1,1],
                     [1,1,5]])
conv_image = cv2.filter2D(image, -1 , kernel)
cv2_imshow(conv_image)


```
